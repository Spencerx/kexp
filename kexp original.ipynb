```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('song_counts.txt', names=['artist', 'title', 'count'])
```

```python
df.head(10)
```

```python
plays_by_artist = df.groupby(['artist']).aggregate(np.sum).sort_values(by='count',ascending=False)
```

```python
plays_by_artist.head(10)
```

```python
tracks_by_artist = df.groupby(['artist']).size().sort_values(ascending=False)
```

```python
tracks_by_artist.head(10)
```

```python
artists_by_title = df.groupby(['title']).size().sort_values(ascending=False)
```

```python
artists_by_title.head(40)
```
